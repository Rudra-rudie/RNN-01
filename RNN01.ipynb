```python
#tokenize
def tokenize(text):
  text = text.lower()
  text = text.replace('?', ' ')
  text = text.replace("'", "")
  return text.split()
```

```python

#vocab
vocab = {'<UNK>':0}
```

```python
def build_vocab(row):
  print(row['question'],row['answer'])
  tokenized_question = tokenize(row['question'])
  tokenized_answer = tokenize(row['answer'])
  merged_tokens = tokenized_question + tokenized_answer

  for token in merged_tokens:
    if token not in vocab:
      vocab[token] = len(vocab)

 # print (tokenized_question, tokenized_answer)
```

```python
#conversion
def text_to_indices(text, vocab):
  indexed_text=[]
  for token in tokenize(text):
    if token in vocab:
      indexed_text.append(vocab[token])
    else:
      indexed_text.append(vocab['<UNK>'])

  return indexed_text
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
class QADataset(Dataset):
    def __init__(self, df, vocab):
        self.df = df
        self.vocab = vocab
    def __len__(self):
        return self.df.shape[0]
    def __getitem__(self,index):
        # self.df.iloc[index] # This line was not doing anything, can be removed or implicitly handled by next lines
        numericalized_question = text_to_indices(self.df.iloc[index]['question'], self.vocab)
        numericalized_answer = text_to_indices(self.df.iloc[index]['answer'], self.vocab)

        return torch.tensor(numericalized_question), torch.tensor(numericalized_answer)
```

```python
learning_rate = 0.001
epochs = 20
```

```python
class SimpleRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim=50, hidden_size=100):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.LSTM(embedding_dim, hidden_size, batch_first=True) # Changed to LSTM
        self.fc = nn.Linear(hidden_size, vocab_size)

    def forward(self, x):
        x = self.embedding(x.long())
        out, (h, c) = self.rnn(x) # LSTM returns (output, (h_n, c_n))
        out = self.fc(h.squeeze(0))
        return out
```

```python
model = SimpleRNN(len(vocab))
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), learning_rate)
```

```python
for epoch in range(epochs):
    total_loss = 0

    for question, answer in dataloader:
        optimizer.zero_grad()

        output = model(question)  # model returns (batch_size, 1, vocab_size)
        loss = criterion(output, answer.squeeze(0))  # output is (1, vocab_size), answer.squeeze(0) is (1,)

        loss.backward()

        optimizer.step()

        total_loss = total_loss + loss.item()

    print(f"Epoch: {epoch+1}, Loss: {total_loss/len(dataloader):.4f}")
```

```python
from torch.nn.utils.rnn import pad_sequence

def collate_fn(batch):
    questions, answers = zip(*batch)
    questions = pad_sequence(questions, batch_first=True, padding_value=0)
    answers = torch.stack(answers)
    return questions, answers

train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
train_data, test_data = torch.utils.data.random_split(dataset, [train_size, test_size])

train_loader = DataLoader(train_data, batch_size=32, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_data, batch_size=32, collate_fn=collate_fn)

for epoch in range(epochs):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for question, answer in train_loader:
        optimizer.zero_grad()
        output = model(question)
        loss = criterion(output, answer.squeeze(1))
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        probabilities = torch.nn.functional.softmax(output, dim=1)
        _, predicted = torch.max(probabilities, dim=1)
        correct += (predicted == answer.squeeze()).sum().item()
        total += answer.size(0)

    train_accuracy = correct / total * 100

    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for question, answer in test_loader:
            output = model(question)
            probabilities = torch.nn.functional.softmax(output, dim=1)
            _, predicted = torch.max(probabilities, dim=1)
            correct += (predicted == answer.squeeze()).sum().item()
            total += answer.size(0)

    test_accuracy = correct / total * 100
    print(f"Epoch {epoch+1}/{epochs} | Loss: {total_loss/len(train_loader):.4f} | Train Acc: {train_accuracy:.2f}% | Test Acc: {test_accuracy:.2f}%")
```
